```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('mobile_price_classification.csv')
```

```python
df
```

```python
col_names = df.columns.values.tolist() 
col_names
```

```python
if 'price_range' in col_names:
    
    col_names.remove('price_range')
     
```

```python
col_names
```

```python
x = pd.DataFrame() 
x = x.reindex(columns=col_names)
x = df.columns 

x 
```

```python
x = df.drop(columns='price_range')
x 
y = df['price_range']
y
```

```python
from sklearn.model_selection import train_test_split 

x_train , x_test , y_test , y_train = train_test_split(x , y , test_size = 0.25 , random_state=42)
```

```python
from keras.models import Sequential , load_model
from keras.layers import Dense , Dropout
```

```python
model = Sequential()
model.add(Dense(units=8 , activation='relu' , input_dim = len(x_train.columns)))
model.add(Dense(units=4 , activation='relu' ))
model.add(Dense(units=1 , activation='sigmoid'))

```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy' , optimizer='sgd' , metrics=['accuracy'])
```

```python
model.fit(x_train , y_train , validation_data=(x_test , y_test) , epochs=100 , batch_size=32)
```
